Methods to Handle missing values

2. Dropping

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
dataset = pd.read_csv('Placement_Dataset.csv')
```

```python
dataset.head() # Display first 5 rows of the dataset
```

```python
dataset.shape  # Get the number of rows and columns in the dataset
```

```python
dataset.isnull().sum()  # Check for missing values in each column
```

```python
# replacing null values with central tendency: 1. mean for continuous data 2. mode for categorical data 3. median for skewed continuous data
# analyse the distribution data in the salary column
fig, ax = plt.subplots(figsize=(8, 6))
sns.displot(dataset.salary)
```

In the abouve graph we can see that the data is right skewed as the distribution is more on one side so we will use median to fill the missing values in salary column instead of mean

```python
# Replacing the missing values in salary column with median values
dataset['salary'].fillna(dataset['salary'].median(), inplace=True) # we use inplace=True to make the changes in the original dataset
```

```python
dataset.isnull().sum()  # Check for missing values again to confirm that they have been handled
```

Dropping Method

```python
salary_dataset = pd.read_csv('Placement_Dataset.csv')
```

```python
salary_dataset.shape
```

```python
salary_dataset.isnull().sum()
```

```python
# dropping the rows with missing values
salary_dataset = salary_dataset.dropna(how='any') # how='any' drops rows with any missing values, how='all' drops rows with all values missing
```

```python
salary_dataset.isnull().sum()
```
